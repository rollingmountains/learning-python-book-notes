```python
# default passing by assignment 
# this is in the definition scope
def func(c, a=10, b=30):
    return a + b + c

# default assignment in header gives the caller an option to skip
# passing the matching arg and it would still work with defaul fallback
func(10)
func(30, 40, 50) # assigns caller's object when passed
```

```python
# function as first class citizen object
def add(*add_iterable):
    return sum(add_iterable)

def multiply(*mul_iterable):
    total = 1
    for item in mul_iterable:
        total = total * item
    return total

def exit_instruction():
    print('Thank you. Come again!')

args = [1,9,3]
dispatch = {
    '+': add,
    '*': multiply
}

commands = {
    'exit': exit_instruction
}


while True:
    choice = input('Enter the operator or enter "exit" to end program: ')
    
    cmd = commands.get(choice)
    
    if cmd:
        cmd()
        break
   
    
    func = dispatch.get(choice)
    
    if func is None:
        print('Please enter a valid operator')
        continue  

    print(func(*args))
    print('Enter exit to end the program')
    



    
```

```python
# domain behaviour function
def do_len(text):
    return len(text)

def do_lower(text):
    return text.lower()

def do_upper(text):
    return text.upper()

def do_title(text):
    return text.title()

# meta command behaviours
def exit_instruction():
    print('Thank you. Come again!')

# domain behaviour or action lookup table
text_ops = {
    'len': do_len,
    'upper': do_upper,
    'lower': do_lower,
    'title': do_title
}

# meta behaviour or command lookup table
commands = {
    'exit': exit_instruction }

while True:
    # user input
    user_text = input('Please enter the string: ')
    command = commands.get(user_text)

    if command:
        command()
        break
        
    user_operation = input('Please enter the operation: ')
    
    operation = text_ops.get(user_operation)

    if operation is None:
        print('Please enter a valid operation')
        continue

    print(operation(user_text))





    
```

```python
def intersection(*iterables):
    result = []
    for item in iterables[0]:
        if item not in result:
            for other in iterables[1:]:
                if item in other:
                    result.append(item)
    return result

intersection([1,2,3,3], [3,4,5])
```

```python
def intersection(*args): 
    parts = []
    result = []
    
    for sequence in args:
        parts.append(sequence.split(' '))
    
    for word in parts[0]:
        # check if item already exists, skip current iteration if exists
        if word in result: 
            continue
        
        for sequence in parts[1:]:
            # if item doesn't exists even in one iterable break out of loop and start the outer loop for next iteration
            if word not in sequence: 
                break
        else:
            result.append(word)
            
    return result

intersection('test in jest', 'best in west')
```

```python
# print has args 'iterable' or object, sep for seperate collection of iterables, end to design end of output, and also option to print or send the output to an external file

# inputs
# collection of iterables, sep value, end value, file value
# import sys to print output in terminal as fallback

import sys

def construct_print(args, sep):
    # work with the args
    if not args:
        return 'sorry there is no input to print. please provide an input'
    
    # design invariant
    # collect output
    output = ''
    first_arg = True 
    for arg in args:
        # make sure first arg doesn't apply sep and normalize all iterables as string before printing
        output += ('' if first_arg else sep) + str(arg)
        # make sure rest of args do implement sep
        first_arg = False
    return output

def print3(*args, **kwargs):
    # collect keyword values
    # design failure to receive kwargs value with a default fallback 
    sep = kwargs.get('sep', ' ')
    end = kwargs.get('end', '\n')
    file = kwargs.get('file', sys.stdout)
    
    if args: 
    # core transformation function
        output = construct_print(args, sep)
        
        file.write(output + end)
        
        file.close()

        print(output + end)
    else: 
        print('Sorry there are no items to print')
  
print3(file=open('print.txt', 'w'), end='\n\n\n')
        
```

```python
# desing a print function that tracks an invalid keyword argument
import sys
def construct_print_str(sep, args):
    parts = []
    for arg in args:
        parts.append(arg)
    return sep.join(parts)

# check for extra keyword args using explicity keyword at the signature 

def print3(first, *args, sep=' ', end='\n', file=sys.stdout):
    pargs = first, *args
    print_string = construct_print_str(sep, pargs)
    output = print_string + end
    file.write(output)
    
    
    
print3('this is whim', 'this is jime', file=open('print.txt', 'w'))
    
    
```
