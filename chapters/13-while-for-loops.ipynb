```python
# traverse a list sequence and change the list as traversing happens
l = [1,2,3,4,5]

for item in l:
    item += 1 # updates only the last of the item and that's not reflected in l
              # this is simply because item is reassigned each time and is actually not inside a container to hold all the values

# this is a good candidate for utilizing access to index using range and len of sequence
for i in range(len(l)):
    l[i] += 1 # incremements all the items by 1

# while loop can also achieve the same result but needs extra counter to keep track of the index
l = [10,20,30,40]
i = 0 # external counter

while i < len(l):
    l[i] += 1
    i += 1 # manual increment of the counter
```

```python
# make the while loop composable using a generator
def read_line():
    while True:
        line = input('>>> ')
        if not line:
            break
        yield line
        
for line in read_line():
    print(line)
```

```python
# desing custom range generator
class CustomRange:
    def __init__(self, limit):
        self.limit = limit
        self.counter = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.counter < self.limit:
            value = self.counter
            self.counter += 1
            return value
        else:
            raise StopIteration

def range_generator(limit):
    counter = 0
    while counter < limit:
        yield counter
        counter += 1
        
for num in range_generator(5):
    print(num)
        
```

```python
# design enumerate using yield for all iterators
def custom_enumerate(iterable, start=0):
    counter = start
    
    if isinstance(iterable, dict):
        iterator = iterable.items()
    else:
        iterator = iterable
        
    for item in iterator:
        yield counter, item
        counter += 1

for i, item in custom_enumerate({'name': 'john', 'age': 40}, start=100):
    print(i, item)
    
for i, value in enumerate({'name': 'john', 'age': 40}.items()):
    print(i, value)
```

```python
# understanding iteration protocol
# a method that returns self, meaning returns itself with no copy but itself

class MyClass:
    def __init__(self, value):
        self.value = value
        
    def get_itself(self):
        return self # returning the object itself as is not the value
    #but the whole object
    
    def __repr__(self):
        return f'{self.value}'
    
c = MyClass(42)
c1 = c.get_itself()
c1.value = 100
c, c.get_itself(), c1, c1.get_itself(), id(c), id(c1), id(c1.get_itself())


```
