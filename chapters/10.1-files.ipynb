```python
# context manager for files
with open('data.txt', 'w+') as f:
    f.write('this is a clumsy world')
    f.seek(0)
    c = f.read()

print(c)
```

```python
# copies and refernces
x = [1,2,3,4]
y = x # this is a reference
x, y, id(x), id(y)

y = x[:] # this is a copy
y[0] = 'a' # this changes or mutates only y and not x
x, y, id(x), id(y)

# using constructor to make a copy
x = [1,2,3]
y = list(x) # this also creates a copy

x, y, id(x), id(y)

# use copy method to create a copy
d = {'name': 'john', 'age': 30}
c = d.copy()

d, c, id(d), id(c)
```

```python
# copy methods only make copies of the high level and leave the neseted objects as references, to solve this we need to make use of deepcopy method from the standard libaray

x = [1,2,[3,[4,6]],5]
y = x.copy()
# y[2] = 3 # this is only changing the refrence at index 2, so while y index 2 points to the new object 3, the x remain unchanged
y[2][0]= 'a' # this directly works on the inside references and mutates the value the refernces point to, hence value of x also changes because it's also pointing to the same thing
y[2][1] = 'b'
x, y, id(x), id(y), id(x[2]), id(y[2])

```
