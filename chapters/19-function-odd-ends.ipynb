```python
# summation using recursion
def my_sum(L):
    print(L)
    if not L:
        return 0      
    else:
        return L[0] + my_sum(L[1:])

r = my_sum([1,2,3,4])
r
```

```python
# summation using ternary operator
def summation(L):
    return 0 if not L else L[0] + summation(L[1:])

summation([1,2,3,4,5])
```

```python
# summation on generalized iterables not just numbers
def summation(L):
    return L[0] if len(L) == 1 else L[0] + summation(L[1:])
summation([1,2,3,4]), summation(['test', 'best', 'jest']), summation((5,6,7)), summation(open('print.txt'))
    
```

```python
# use unpacking to work with the items in the collection
def summation(L):
    first, *rest = L
    return first if not rest else first + summation(rest)

summation([1,2,3,4]), summation(['test', 'best', 'jest']), summation((5,6,7)), summation(open('print.txt'))
```

```python
counter = 0
def sumtree(L, trace=False):
    global counter 
    counter += 1
    print(f'counter {counter}')
    print('@@@@@@@')
    tot = 0
    for x in L:
        if not isinstance(x, list):
            tot += x 
            if trace: 
                print(x, end=(','))
        else:
            tot += sumtree(x, trace)

    return tot

sumtree([1,[2,[3,4],5],6], trace=True)
```

```python
# indirect recursion
def mysum(L):
    if not L: 
        return 0
    return nonempty(L)

def nonempty(L):
    return L[0] + mysum(L[1:])

mysum([1,2,3,4,5])
    
```
