```python
# string expression formatting

'this is a text' #  this is a string literal
'this is a %s' #  this is a format string literal

# expression format full syntax 
'this is a %s' % 'expression format' # --> 'this is a expression format'

'%s' # % modulo indicates start of the conversion target and s is a typecode which expects a string as replacement value on the right side of the operator % 'expression format'

'%d-%s-%f' % (3,'three',3.0) # --> '3-three-3.000000'
# seen above, we can used different type to be specific about the object type we expect in the tuples however since the whole is a string at the end, just %s also works the same
'%s-%s-%s' % (3,'three',3.0) # --> '3-three-3.0' since the string is as is, it prints the float object 3.0 as is and not as 3.000000

# using options in the conversion type syntax for width, padding and precision plus keyname during dictionary

# %[keyname][flags][width][.precision]typecode
# %-6d -> - is a justification at end saying left aling the object, 6 is the width of the total decimal number including width 
# %06d --> 0 is padding to be used with the decimal and 6 is the width of the total decimal number including width
x = 1234
'%-6d' % x # --> '1234 ' original decimal width is 4 + 2 = 6 total width, and 2 space for justification at the end
'%6d' % x # --> '1234 ' original decimal width is 4 + 2 = 6 total width, and 2 space for justification at the begining. meaning right align the object,  becuase of postive number 6
'%06d' % x # --> '001234' original decimal width is 4 + 2 = 6 total width, and 2 0's in the front for padding. by default the padding makes the object right align and there is no option to make the padding on the right and make the object left align in expression format. you can only # using other string methods like ljust
str(x).rjust(6, '0')
```

```python
# expression formatting for floating precision

x = 1.23456789


x, '%f--%e--%.2f--%g' % (x, x, x,x)

# using left alingment with - and also width in the option along with precision for a float number
'%-6.2f' % x # --> '1.23  '
'%05.1f' % x # -->'001.2' padding in the start and str right alignment width size 5 and .1 precision
'%+06.2f' % x # --> '+01.23' using + to indicate positive number explecitly and then using 0 as padding and 6 as width, .2 as precision. 
'%-05.1f' % x # --> '1.2  ' - symbol works as left alignment and no math symbol works as right alignment using + is used as is. using -05.1f would simply use left alignment and consider 05 as 5 which is then taken to be width size

```

```python
# pass the options to the conversion target dynamically in the tuple

'%.*f' % (4, 1.2345678) # 4 is precision and 1.2345678 is value

# pass other options too but in the correct order of options 
# %[keyname][flags][width][.precision]typecode

'%*.*f' % (10, 8, 1.2345678) # --> '1.23456780' => 10 = total width and 8 = precision

'%*.*f' % (11, 8, 1.2345678) # --> ' 1.23456780' => 11 = total width and 8 = precision, and right justifies the string befault justify when no flags passed. because of length of the width > actual length of the string

'%-*.*f' % (11, 8, 1.2345678) # --> '1.23456780 ' => 11 = total width and 8 = precision, and left justifies the string due to -. because of length of the width > actual length of the string

# padding with 0's at the begining and end can be achieved by replacing '-' or the default positive by using 0

'%0*.*f' % (11, 8, 1.2345678) # --> '01.23456780' => 11 = total width and 8 = precision, and right justifies and pads with 0 on the right side. because of length of the width > actual length of the string

'%-0*.*f' % (11, 8, 1.2345678) # --> '1.23456780 ' => 11 = total width and 8 = precision, and left justifies the string due to -. because of length of the width > actual length of the string. The 0 isn't considered because the - wins and treats 0 as the trailing of width 11 as 011 == 11. So essentially padding with 0's at the end of the string can't be achieved using the expression format and need to rely on the ljust

'1.2345678'.ljust(11, '0') # --> '1.234567800'
```

```python
# expand string format expression to support dictionary based values

# format string literal expands the coversion target syntax to support dict based values
'My name is %(name)s. I am %(age)s years old' % {'name': 'Pat', 'age': 35} # --> 'My name is Pat. I am 35 years old'

# use a single dict key in mutiple places within in a string
'%(animal)s are very shy. %(animal)s loves to drink milk' % {'animal': 'Cat'} # --> 'Cat are very shy. Cat loves to drink milk'

# make use of built in vars function to organise variables as dict in place
name = 'Pat'
age = 30


'My name is %(name)s. I am %(age)s years old' % vars()


```

```python
# string method format
# basic syntax
'This is a {} {}'.format('string', 'literal') # relative position --> 'This is a string literal'
'This is a {0} {1}'.format('string', 'literal') # absolute position --> 'This is a string literal'
# absolution position need not be in order in the placeholder
'This is a {1} {0}'.format('literal', 'string') # --> 'This is a string literal'

# keyword value source
'This is a {first} {second} supporting keyword.'.format(first='string', second='literal')

# combining keyword and absolute position
'{first} {second} a {0} {1} supporting mixed source'.format('string', 'literal', first='This', second='is')
```

```python
# format method doesn't allow mixing of automatic {} and manual position {1} but manual and keyword {first} for simplicity and no real world use case however it does allow mixing manual approaches 

'{} {} {} {first} {second} {}'.format('This', 'is', 'a', 'mixing', first='string', second='literal') 
```

```python
# format method allows all types of objects in the object source
'{0}, {1}, {2}, {3}, {name}, {4}'.format(10, 'ten', 10.00, [1,2,3,4], (1,2,3,4,5), name='pat')
```

```python
# format method also allows accessing attributes in the conversion target inside {}

import math as m

'{0.pi}'.format(m) # use absolute position or keyword to access the object inside the comparison target
'{name.pi}'.format(name=m) # use the keyword to access the object and its attributes 

'{lst[0]}'.format(lst=[1,2,3]) # use the offset or the indicies of the object to access its value

'{dict[name]}'.format(dict = {'name': 'pat'})
```
