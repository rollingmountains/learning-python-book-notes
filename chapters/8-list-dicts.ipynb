```python
# List basic operation
[1,3] + [1,3] # -> [1, 3, 1, 3]
[1,3] * 2 # --> [1,3,1,3]

# Mixed collection of objects with basic operation
# Behaves the same as same type collection
# ['hello', 123, [3,4,5], {'name': 'pat'}] + ['hello', 123, [3,4,5], {'name': 'pat'}]
['hello', 123, [3,4,5], {'name': 'pat'}] * 3

# Basic operation on mixed types as operand is not supported
# [1,2,3] + 'hello' -> throws an error can only concatenate list to list
```

```python
# list creation
lst = 1,2,3,4,5,6 # -> [1,2,3,4,5,6]
lst = 'hello' # -> ['h', 'e', 'l', 'l', 'o']
lst = {'name': 'pat', 'age': 40} # -> ['name', 'age'] converts the keys of list as the collection of items
type(list(lst)), type(lst.keys()) # -> (list, dict_keys)
# dict.keys(), dict.items() are view only they can be iterated over or check membership or compared with other dict keys but not indexed or subscripted
# to convert dict keys or items into subscriptable pass it into list constructor

k = list(lst.keys()) # -> ['name', 'age'] collection of keys as string objects
k[1] # -> 'age'

i = list(lst.items()) # -> [('name', 'pat'), ('age', 40)] collection of key-value pair as a tuple objects
i[0], i[0][1] # -> (('name', 'pat'), 'pat') access using indexing just like done on any sequence object in python
```

```python
# value comparison operators on flat list
[1,2,3] == [1,2,3], id([1,2,3]) == id([1,2,3]), id([1,2,3])
# (True, True, 4464848832) all refrences the same object in the memory 

a = [1,2,3]
b = [1,2,3]
a == b, id(a) == id(b), id(a), id(b), id(a[0]) == id(b[0])
# -> (True, False, 4465254336, 4465076864, True)
# here a and b reference different objects in the memory, however the collection objects inside these different objects refrence to the same integer objects in the memory. hence False when the list objects are compared but True when the items of same offset are compared
```

```python
# value comparison operator on neseted list
a = [1,2,[3,4],5]
b = [1,2,[3,4],5]

a == b, id(a) == id(b), id(a), id(b), id(a[2][1]) == id(b[2][1])
# (True, False, 4465251584, 4464868672, True)
# here a and b reference different objects in the memory, however the collection objects inside these different objects refrence to the same integer objects in the memory. hence False when the list objects are compared but True when the items of same offset are compared
```

```python
# ordering operator on flat list
a = [1,2,3]
b = [1,2,3]

a == b, a < b, id(a) == id(b), a <= b
# -> (True, False, False, True)
# == compared each value of same offset from both the list
# < compared each value of same offset from both list 
# <=compared each value of same offset from both list

a = [1,2,3]
b =[1,2,2]

a == b, a > b, id(a) == id(b)
# -> (False, True, False)
# the last offset value of list 'a' is bigger than last offset value of list 'b' hence in ordering list 'a' comes after list 'b' so 'a' is greater than 'b'

a = [1,3]
b = [1,2,1]

a == b, b > a
```

```python
# iteration in list
# manual for loop
result = []
for x in [1,2,3]:
    result.append(x*4)  
result

# list comprehension is concise and efficient way to rewrite manual for loop
# list comprehension removes the need to manually set an empty list and the append operation to add the members into the new list object
# list comprehension applies an expression to every members in the collection or sequences and wraps them in a square bracket to create list object
result = [x * 4 for x in [1,2,3]]
result

# list comprehension can be extended by adding either a filter condition or by applying map built in methods 
result = [x * 2 for x in [1,2,3] if x > 1] # applies filter on the member before the expression is applied on the sequence. if the filter results in False then the expression isn't applied on the member, only when the filter is True on the item the expression is applied and added to the list
result
```

```python
# contrasting list comprehension and map built-in function
def square_root(x):
    return x * x

result = [square_root(x) for x in [1,2,3]]
result

result = map(square_root, [1,2,3])

# lst_result = list(result)
# lst_result

result = (square_root(x) for x in [1,2,3])
result
next(result)
next(result)
next(result)

result = map(square_root, [1,2,3])
next(result)
```

```python
# another way to remove item from dictionary is using the popitem method

# the order of preservation in dictionary is from left to right, and any new key values are added at the end, the popitem function removes the last collection item following the LIFO convention

d = {'name': 'john', 'gender': 'male', 'age': 30}
d.popitem(), d.pop('name'), d # (('age', 30), 'john', {'gender': 'male'})

# popitem removes the last item and returns the key value pair
# pop function removes the key passed as the argument(key can be anywhere in the dict) and returns the value mapped to that key
```
