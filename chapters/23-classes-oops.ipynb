```python
# function a reusable code
# customization needs changes in place or creation of new function 
def greet(name):
    return f'Hello, {name}'
# customization with new function 
# this is customization with redundancy
def loudgreet(name):
    return f'Hello, {name.upper()}'

# class customizable code
# customization doens't need to change in place but uses inheritence
class Greeter:
    def greet(self, name):
        return f'Hello, {name}' 

# here is the customization
# LoudGreeter inherits from Greeter, uses its greet function and extends its functionality
# this gives customization without redundancy
class LoudGreeter(Greeter):
    def greet(self, name):
        return super().greet(name.upper())
    
# class are typical open for extension and closed for modification. Open and Close principle

g = Greeter()
lg = LoudGreeter()

g.greet('alice'), lg.greet('bob')
```

```python
# a class is just a package of functions and that use and process the built-in objects provided by python
import math

# built-in objects provided by python; we have been using until now
name = 'Alice' # a str built in object
scores = [60,90,85] # a list built in object
email = 'alice@test.com' # a list built in object

# a class can now design functions in a package such that these built-ins are used and processed within it

class Student:
    def __init__(self, name, scores, email):
        self.name = name # built in str bound to variable/name
        self.scores = scores # built in list of built in int is bound to a variable/name
        self.email = email # built in str bound to a variable/name
        
    def average(self):
        return round(sum(self.scores) / len(self.scores)) # processes the built in list using sum and len, divides them for calculation, basically processing
    
    def greet(self):
        return f'Hello, {self.name}' # using the built in str to generate a greet message
    

class GradStudent(Student):
    def average(self):
        return super().average() + 5

g = Student(name, scores, email)


gs = GradStudent(name, scores, email)
g.name, g.scores, g.email, g.greet(), g.average(), gs.greet(), gs.average()
```

```python
# example of a GUI with window and widgets
# A GUI can be de-composed into windows, widgets, buttons, labels, fancybuttons, menu etc

# seam that keeps all the items(button, lables, fancy-button) inside the window is a widget

# these items can be then de-composed with a shared contract since they are part of a widget

# make a widget object that sets the contract that each item inside a widget can actually 

class Widget:
    def draw(self):
        pass
    
class Button(Widget):
    def draw(self):
        print('drawing button')
    def color(self):
        print('set the button color')
        
class Label(Widget):
    def draw(self):
        print('drawing label')
        
class FancyButton(Button):
    def draw(self):
        super().draw()

# an object that composes all these de-composed object into a single unit
class Window(Widget):
    def __init__(self, name, widget_items):
        self.name = name
        self.widget_items = widget_items
        
    def draw(self):
        print('starting the drawing')
        print('@@@@@@@@@@@@@@@@@@@@')
        for widget in self.widget_items:
            widget.draw()
        print('@@@@@@@@@@@@@@@@@@@@')
        print('drawing is complete.')

b = Button()
l = Label()
fb = FancyButton()
w = Window("canvas",[b,l,fb] )
w.draw()

            
```

```python
# Feature 2

# Customization via Inheritance

class Car:
    def __init__(self, name, model):
        self.name = name
        self.model = model
    
    def engine(self):
        print('start the engine')

class TurboCar(Car):
    def __init__(self, name, model, turbo_charge):
        super().__init__(name, model)
        self.turbo_charge = turbo_charge
        
    def speed_up(self):
        super().engine()
        print('turbo charge applied')
        
tc = TurboCar('Torque', 'Sonic', 1000)

TurboCar.mro()
[x for x in dir(tc) if not x.startswith('__')] # all the properties and methods new object can access including its parent and super parent
vars(tc) # new objects properties alone, methods live in its parent class
[x for x in dir(TurboCar) if not x.startswith('__')] # has attribute method speed_up and initial/static attribute turbo_charge
vars(TurboCar)
# vars(Car) # has attribute method speed_up and initial/static attribute turbo_charge
```

```python
# continuation from above cell
# now Animal class can override the __init__ method that belongs to class type, this overriding is called as 'operator overloading', here the type's init operator is being overridden by Animal's init
class Animal:
    # init is overriding type's init
    def __init__(self, name, color):
        self.name = name
        self.color = color
        
    def speak(self):
        return 'The Animal object is speaking'
    
    # operator overloading the __add__ method 
    def __add__(self, other):
        return self.name + ' speaking ' + other
    
cat = Animal('mili', 'black')
cat.name, cat.color, cat.speak()

cat + '...test ...test...' # + is the interface to access or invoke the __add__ method on cat

# when python sees + it does this
cat.__add__('...test...test...')

# Though __add__ exists in the primordial object class, python looks for this method starting from the object --> parent class Animal --> all the way up to primordial object. it stops the search as soon as it finds one
```

```python
# operator scheme overloading
# take the operator & 
class SetIntersection:
    def __init__(self, value):
        self.value = value
    
    def setintersect(self, set2):
        return self.value & set2
    
    def __and__(self, set2):
        return self.value & set2
s= SetIntersection({1,2,3,4,5,6})

s1 = {1,2,3,4,5,6}
SetIntersection.setintersect(s, {3,4,5,6,7,8}), s.setintersect({3,4,5,6,7,8}),s & {3,4,5,6,7,8}, s1 & {3,4,5,6,7,8}


```
