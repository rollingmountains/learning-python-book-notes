```python
# a variable in python is only a name or simply put reference to the memory address where the value is stored in RAM
x = 'string object' # the string literal string object is assinged to the variable name x, x points to the memory address of the string object
id(x) # x points to the memory address (4370840944) of the name x, this memory address actually holds the value, its type and the reference countof the string object
```

```python
# shared reference or shared immutable object 
x = 100 # 100 int object assigned to x
y = x # x which resolves to int object 100 is assigned to y
id(x) == id(y) # both x and y refer to the same memory address of int object 100
x is y
```

```python
# interning of small integers
x = 10
b = 10
x is b # small int so it's true
x = 900
b = 900
x is b # big int beyond 256, so it's false
x = 'name'
y = 'name'
x is y # sometimes python does intering for short string objects this gives true
x = 'this is a beautiful world'
y = 'this is a beautiful world'
x is y # for this long string object python does NO intern, hence it's a False
```

```python
# shared reference immutable object, when name points to another object
x = 10
y = x
x += 10 # now x is assinged to another new int object via expression
x is y # this is False because x no longer points to same object as y
y # y continues to points to the same object 10, change in the x assignment has no impact on the object y points to 

# this shared refernece makes an object to be passed around different parts of the program, in different functions without impacting the original object value
# this is beneficial especially in pure functional programming, where a function is expected to return an expression or an object with no side effects that is make no changes to the state of the object it received 
```

```python
# shared object and mutable objects, changes made in place
x = [1,2,3,4]
y = x
x is y # same as mutable object here both x and y point to the same memory address

x.append(5) # add element to the object x points to, essentially, update the state of the list object, making change in place
x # x points to the new modified object address
y # also points to the new modified object since it was sharing the reference with x

x is y # this still True because both refer to the same object in memory though changes to the object can be applied either via x or y
y.append(6) # apply changes to the list via y name
y # this reflects the list with new item object 6
x # x also reflects the changes since it points to the same object in the memory
y is x # obviously this becomes True

x = 'this is good' # refer x name to a different object type
y,x # x and y point to diffferent objects in memory, y points to the list object and x pointst to the string object
x is y # this is obviously False
```

```python
# retain original values of the mutable shared reference object
# make a copy of the object before changing in place

x = [1,2,3,4,5]
y = x[:]
x,y
x is y
```

```python
m = 1,2,3
l = m

# comparing values
l == m
# check if the objects are same
l is m

m = 1,2,3 # a list object
l = 1,2,3 # another list object

m == l # are values equal -> True
l is m # are the objects equal -> False, since m and l point to different objects in the memory though both the objects have same value


```

```python
# shared reference, immutable objects and equality
x = 10
y = x

x == y # check value -> True
x is y # check same objects -> True

x = 100
y = 100

x == y # check value -> True
x is y # check object -> True, though looks like both names point to different object in memory, in reality python caches small integers so under the hood 100 object assinged to both x and y is the same in memory 
```
