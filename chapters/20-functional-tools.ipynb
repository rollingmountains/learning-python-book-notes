```python
# function annotation and using decorator to work with the annotation
def annotate(func):
    for key, value in func.__annotations__.items():
        print(key, '=>', value)
    def extra(*args):
        total = func(*args)
        total += 20
        return total
    return extra


@annotate
def add(a:'first_arg', b:'second_arg') -> int:
    return a + b

add(1,2)
```

```python
# more concrete example
def wrapper(f):
    def proxy(*args):
        print('I am the proxy')
        return f(*args)
    return proxy

@wrapper
def func(x, y):
    print('running actual function object', x, y)
    
f = func(1,2)
w = wrapper(f)
    
    
    
```

```python
# map function from scratch 
def mymap(func, iter):
    for i in iter: 
        yield func(i)
    
    

def inc(x):
    return x + 30

iteration = [1,2,3]

update = mymap(inc, iteration)

# build list in a single line of code
list(mymap(inc, iteration))

# use inline function lambda
list(mymap(lambda x: x + 20, iteration))

# simulate the map functionality using list comprehension 
funcs = [(lambda item=item: item + 50) for item in iteration]
[f() for f in funcs]
```

```python
# classic reduce function
r = reduce(lambda x, y: x + y, [1,2,3,4,5])


# build a reduce function from scratch
from operator import add, mul
iterables = [1,2,3,4,5]
def myreduce(iterables, operation):
    operators = {
        'add': add,
        'mul': mul,
    }
    
    func = operators[operation]
    accumulator = 0 if operation == 'add' else 1
    
    for item in iterables:
        accumulator = func(accumulator, item)
    return accumulator

myreduce(iterables, 'mul')

# how to make the reduce function generic 
```
