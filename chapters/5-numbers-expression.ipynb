```python
# Python expression operators 
# expression is a combination of names/objects/value and operators example x + y
# there will also be combination of multiple operators like (x * y) + (z * z)

# Mixed operator precedence
# order is higher precedence by operators at the lower part of the operator table cited in the book, and when operators are in same row then it moves left to right. however there is an exceptions in side ways direction, exponentiation move from right to left

# we can control the default precedence set by python by using a technique called grouping, it's done by using parantheses

(10 + 20) * 30 

# here the + operator takes precedence over * because the sub-expression it holds is wrapped inside a parantheses effectively grouping

# Number types have an order too in terms of complexity, for example integer < float < complex number
3 + 3.90 # python converts int 3 into float and performs addition
3 == 3.0 # this is true because 3 gets converted over to 3.0 and the equality check happens
3 > 3.0  # this returns false

```

```python
# Floating point equality
# when comparing floating point objects or expressions for equality it's always advisable to use the comparison after rounding, fixed decimal parts both both operands, or using the math module's isclose method so that there are no suprises 
# the reprsentation of a decimal number or floats in binary the computer's storage system can't truly represent it, it goes on repeating so it has to be truncated to represent properly
# when this truncate happens there is a tiny round up that happens which unfortunately is no same when comparing with the literal decimal number which is mathematically equal to the expression
0.1 + 0.2 == 0.3 # --> this is Flase here the binary of 0.1 + 0.2 is a a bit different from 0.3 hence though they are mathematically equal they are not binarically equal hence it's a false
# but when compared using isclose it makes sure it doesn't evaluate upto to the last bits 

import math
math.isclose(0.1 + 0.2, 0.3) # --> this is True! hurray
```

```python
# Division
10/2 # true divsion keeps the reminder always for all types
10//2 # --> 5 - floor division truncates the reminder always but the result object depends on the operand objects
10 // 2.0 # --> 5.0 - floor division truncates the reminder always but the result object depends on the operand objects
10 % 3 # --> 1 gives the reminder of the division always, and the result numeric object depends on either of the operand
10 % 3.0 # --> 1.0 gives the reminder of the division always, and the result numeric object depends on either of the operand
divmod(10,3) # (3,1) --> this gives both the part the divison and also the reminder part in a tuple

# division using positive and negative type
5/2, 5/-2 # (2.5, -2.5) --> keeps the reminder always
5//2, 5//-2 # (2, -3) --> floors to the greatest integer that is equal to or less than the result
5/2.0,5/-2.0 # (2.5, -2.5) --> keeps the reminder for float operands too as expected
5//2.0, 5//-2.0 # (2.0, -3.0) --> retains the format of float object since one of the operand is float

# curious case of truncate
import math
math.trunc(5/-2), math.trunc(5/2) # (-2, 2) --> for negative this truncates towards the zero and not floor similar to floor division and this is a problem when it comes to negative number as it floors towards zero
# 2 --> for positive since it truncates and by desing moves towards the next greatest integer when the operands are positive then its behaviour resembels floor division
```

```python
# using set on database rows
engineers = {'bob', 'pat', 'ann', 'sue'}
managers = {'sue', 'tom'}

# membership operation in
'pat' in engineers # True

#engineers and managers using AND
engineers & managers # sue

# engineers and managers full list using OR
engineers | managers # removes duplicate sue and creates a combined set

# engineers who are not managers
engineers - managers # use minus to remove common name from set A that is in set B

# Are all managers engineers?
engineers > managers # false tom is manager but not engineer

# Are these people all engineers 
{'pat', 'ann', 'sue'} < engineers

# Are people superset of managers, meaning do managers come from the list of engineers and managers itself?
(managers | engineers) > managers

# Who are all have just one role and not both
# use exclusive or XOR
engineers ^ managers # sue is remove since sue is both manager and engineer


```
