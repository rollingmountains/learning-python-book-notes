```python
# repetation adds one level deeper
x = [1,2,3]
y = x * 4 # this is a concatenation of list to itself 4 times
# [1,2,3] + [1,2,3] + [1,2,3] +[1,2,3]
z = [x] * 4

x[0] = 'a' # here both x and z are refering to the same object hence change at x will also reflect in z

m = [list(x)] * 4 # list constructor creates a distinct copy of x and repeats it 4 times, meaning all the object each of the indexes in m points to same reference
x[1] = 'b'
m[0][2] = 'c'
x, m, id(x), id(m), id(m[0]), id(m[1]), id(m[2]), id(m[3])

# use range and loop to create copies within m
n = [list(x) for _ in range(0,4)]
n[0][1] = 'z' # changes only the first index and rest of the indexes are untouched
n, id(n[0]), id(n[1]), id(n[2]), id(n[3])
```
