```python
# literal structure
# tuples are immutable unlike the list
t = (1, 3, [4,5])
t
```

```python
# Tuple in action
# sequence methods
t = (1,6) + (4,5) # concatenation
x = ([1,5], {'name': 'bala'}) * 2 # repetition
y = ('hello world', 10, [1,2], {'name': 'john'}) 
y[0], y[-1], y[:2], id(y), id(y[:2]) # indexing and slicing

d = {'name':'john', 'age': 40}
m = (*[1,2,3], {**d}) # list iterable unpacking and dict iterable unpacking
m

```

```python
# Tuple specific methods
t = (1,1,2,3,[5,6])
t.index(2) # index of a given object
t.count(1) # count gives number of occurances of an object
```

```python
# sorting in tuples
t = (11,2,3)
tuple(sorted(t, reverse=True)) # sort using built in function reversed

n = ''.join(sorted('back'))
n

```

```python
# tuple vs dict vs list when keeping records
# tuple and list can keep records but with position to access them and additional responisbility to remember those positions

t = tuple(['john', 40, ['dancing', 'singing']])
t, t[0], t[2]

# dict has advantage over list or tuple because it uses lable to name the record
d = {'name': 'john', 'age': 40, 'hobbies': ['dancing', 'reading']}
d['name'], d['hobbies']

# can convert part of dict into tuple
t = tuple(d.values())
t

# convert the whole dict into a tuple with pairs
t = tuple(d.items())
t[0][0], t[0][1]

# these conversions however don't give an object that provides object that gives positional and named access
```
