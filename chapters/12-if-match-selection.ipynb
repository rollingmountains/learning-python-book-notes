```python
# using dictionary as mutiple choice selection
choice = 'linux'
D = dict(macOS=1992, android=2001, linux=1995)
D.get(choice, 'bad choice') # using get method to pass default value when choice doesn't exist in the list
```

```python
# use simple if statement to make the choice
choice = 'windows'
if choice == 'macOS':
    print(1992)
elif choice == 'andriod':
    print(2001)
elif choice == 'linux':
    print(1995)
else:
    print('bad choice')
```

```python
# use match case for the same use case
choice = 'macOS'
match choice:
    case 'macOS':
        print(1992)
    case 'linux':
        print(1995)
    case 'andriod':
        print(2002)
    case _:
        print('bad choice')
```

```python
# mix of dict and if condition
branch = dict(macOS=2001, linux=1995, android=2005)

choice = 'windows'

if choice in branch:
    print(branch[choice])
else:
    print('bad choice')
```

```python
# use try catch
branch = dict(macOS=2001, linux=1995, android=2005)

choice = 'windows'

try:
    print(branch[choice])
except KeyError as e:
    print(f'{e} key doesn"t exist. Pass a valid key name.')
```

```python
pattern = {'name': 'john'}
match pattern:
    case 'python':
        print('this is a string object')
    case 100:
        print('this is a number object')
    case [1,2]:
        print('this is a list object')
    # case {'test', 'test1'}:
    #     print('this is a set object')
    case {'name': 'john'}:
        print('this is a dict object')
    case (100, 200):
        print('this is a tuple')
    case _:
        print('this does not exists')
```
