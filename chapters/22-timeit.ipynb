```python
import timeit

r = timeit.repeat(stmt='[x ** 2 for x in range(1000)]', number=100000, repeat=10)
r
```

```python

from functools import partial
def test():
    return [x ** 2 for x in range(1000)]
timeit.repeat(lambda : test(), number=20, repeat=10), timeit.repeat(partial(test), number=20, repeat=10)
```
