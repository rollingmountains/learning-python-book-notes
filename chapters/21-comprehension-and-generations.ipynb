```python
# nested list comprehension 
# but first classic nested for loop
# pair numbers from two iterables based on their odd or even nature
result = []
for x in range(5):
    if x % 2 == 0:
        for y in range(5):
            if y % 2 == 1:
                result.append((x, y))

# derive the same result using nested list comprehension 
l = [(x,y) for x in range(5) if x % 2 == 0 for y in range(5) if y % 2 ==1]
result, l
```

```python
#  Generator function 
def squares(n):
    for item in range(1, n):
        yield item ** 2 # yield turns the func into a generator object
        
g = squares(4)
g # here g is a generator object that also an iterator with iter and next implicity applied
next(g), next(g), next(g)
```

```python
# sub generator functions
def greet(s):
    yield f'Welcome, {s}!'
    yield f'Namaskara, {s}'

def howru(s):
    yield f'How are you doing, {s}?'

def wish(s):
    yield f'Wish you a good day, {s}.'

# orchestrator function 
def orchestrator(s):
    print(type(greet(s)))
    yield from greet(s)
    yield from howru(s)
    yield from wish(s)
    
# caller function
r = orchestrator('bala')
type(next(r)), type(next(r))
```

```python
# regular yield structure 
def my_yield(n):
    for item in range(n): yield item # yielding value right here
    for item in map(lambda x: x ** 2, range(n)): yield item # yielding value right here
r = my_yield(5)
# for item in r:
#     print(item)
    
# yield extensio with from clause
def my_yield(n):
    yield from range(n) # delegating the yield task to a sub genertor
    yield from map(lambda x: x ** 2, range(n)) # delegating the yield task to a sub generator

r = my_yield(5)

' : '.join(str(item) for item in r)# this is a generator expression


    
```

```python
# chainining of generators, multiple chains
# bottom generator that actually does the yielding 
def bottom(n):
    yield n * 2
    yield n ** 2
    
# second level delegation that delegates lazy yielding to bottom generator
def second_delegate(n):
    yield from bottom(n)
    
# first level delegation that delegates lazy yielding to second level generator
def first_delegate(n):
    yield from second_delegate(n)

r = first_delegate(5)
list(r)
```

```python
# path: /Users/balapaudel/programming/learning-python

import os

# function to fetch file paths in a directory`
def walk_files(directory):
    for entry in os.scandir(directory):
        if entry.is_file():
            yield entry.path
        elif entry.is_dir():
            yield from walk_files(entry.path)
   
# function to fetch one file at a time, and its line one at a time         
def read_lines(paths):
    for path in paths:
        with open(path, encoding='utf-8', errors='ignore') as f:
            yield from f
            
# process each line for a certain condition
def grep(pattern, lines):
    for line in lines:
        if pattern in line:
            yield line
            
files_path = walk_files('/Users/balapaudel/programming/learning-python/chapters')
lines = read_lines(files_path)
matches = grep('def ', lines)

len(list(matches))

```

```python
# genereator expression vs map
g = list(x ** 2 for x in [1,2]) # inline transformation no func all
m = list(map(lambda x: x **2, [1,2])) # on main map call + repeated lambda call
g, m
# in the above situation gen expression has advantage over map

g = list(abs(x) for x in [1,2,3]) # a python generator frame for loop and then apply c comipled abs
m = list(map(abs, [1,2,3])) # a C compiled map function and apply c compiled abs function 
g,m
# in the above scenario map has advantage becuase it uses faster and effecient C compiled functions and no python level execution whereas gen exp uses python level generator frame before using the C compiled abs function so map is better here
```
