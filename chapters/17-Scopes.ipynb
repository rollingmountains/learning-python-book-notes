```python
import threading
import time
import random

counter = 0

def worker():
    global counter
    for _ in range(10_000):
        # Force interleaving between threads
        current = counter
        time.sleep(random.random() * 0.00001)
        counter = current + 1

threads = []

# Start 10 threads
for _ in range(10):
    t = threading.Thread(target=worker)
    t.start()
    threads.append(t)

# Wait for all threads to finish
for t in threads:
    t.join()

print("Final counter value (should be 100000):", counter)
```

```python
# set up

handlers = {} # a dict to store all user action and their function

# a function to register user action to the handler's storage 
def register_handler(user_action, func):
    handlers[user_action] = func
             
# a function to call the user action function when user performs the action
def trigger(user_action):
    print(f'Calling {user_action} function')
    handlers[user_action]()
             
# a function factory that generates different user action function
def make_handlers(message):
    def handler():
        print(f'Executing user action function: {message}')
    return handler
             
# set of user action, to simulate and see the setup in action
user_actions = ['save file', 'open file', 'edit file', 'export file']


# Executing the setup
# create, register and store the user action and their function 
for action in user_actions:
    handler = make_handlers(action)
    register_handler(action, handler)
             
# call the user action function based on the user action
for action in user_actions:
    trigger(action)
```

```python
# nonlocal changeable per call and sticks to LEGB
# accessing global for persistent data will result in 
# reference to a single object from all the call to the outer function
# unlike nonlocal that gets a fresh copy of enclosing name to the inner
# function
def counter_maker(start):
    global state # creates state name in global namespace
    state = start
    def count(name):
        global state # references state name in global namespace
        state = state + 1 # changes state name in global namespace
        return name, state
    return count

setup = counter_maker(10)
second = counter_maker(10)
setup('first'), setup('first'), setup('first'),setup('first'), setup('first'), setup('first')
second('second'), second('second')

```
