```python
# Interactive loop with new and standard way of writing statments
while True: # ":" tells interpreter that this begins a code block
    reply = input('Enter text: ')
    if reply.lower() == 'stop': break # two statement in same line however second stmt is just a simple expression
    print(reply.upper())
```

```python
# list of target assignment with RHS an iterable like string, list, dict and tuple
# the LHS and RHS should match, RHS unpacked objects is mapped to the LHS names from left to right
a, b, c = 'abc'
a, b, c = ['test', 'best', 'jest']
a, b, c = {'name': 'john', 'age': 40, 'hobby': 'dance'}
a, b, c = 'jan', 'feb', 'march'
a, b, c
```

```python
# unpacking on the LHS
a, *b = 1,2,3,4,5 # *b in the LHS means collect or pack rest of all the items from LHS
a, *b, c = 1,2 # python does single mapping first and whatever is left is mapped to the *b name, when there is no item left them it assings empty list
# a, *b, *c = 1,2,3 # --> multiple starred assignment is not allowed
a, b, c
```

```python
# unpacking in the RHS
lst = [1,2,3,4]

a, b, *c = [*lst] # the starred object in the right should be iterable, with no container the lst unpacks as 1,2,3,4 this isn't an iterable it should then either be wrapped in a list or tuple literal
a, b, *c = tuple(lst)
a, b, c
```

```python
# stars in function header and function calls
def display_items(*args, **kwargs):
    print(args)
    print(kwargs)

L = [1,2,3]
D = {'a': 'one', 'b': 'two'}
display_items(*L, **D)
```
