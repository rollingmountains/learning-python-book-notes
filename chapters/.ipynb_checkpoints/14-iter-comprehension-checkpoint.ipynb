```python
# iteration protocol under the hood for loops
l = [1,2,3,4]
# for calls the iter on l and l returns iterator
# for calls next on the iterator until the list is exhausted
# the loop gracefully exits
e = []
for item in l: 
    e.append(item ** item)
e
```

```python
# iterator protocol utilized by list comprehension
l = [1,2,3,4]
c = [item ** item for item in l]
c
```

```python
# iterator protocol utilized by generator function
def expo(lst):
    for item in lst:
        yield item ** item

generator_instance = expo(l)
g = list(generator_instance)
g

    
```

```python
# strange case of range
# range is also an iterator factory and it's only an iterable
r = range(3)
# for i in r:
#     print(i)
# for i in r:
#     print(i)
# range doesn't get exhausted because it can be iterated multiple 
# times on the same range object

# range can also be indexed and reversed
r = reversed(range(5))
for i in r:
    print(i)
l = reversed([10,20,30])
for i in l:
    print(i)
```

```python
# traverse a sequence without iteration protocol
counter = 0
lst = [1,2,3]
while counter < len(lst):
    item = lst[counter]
    counter += 1
    print(item, end=' ')
```

```python
# traverse a sequence with for loop and use no iteration protocol 
lst = [1,2,3]
i = 0
for _ in lst:
    print(lst[i], end=' ')
    i += 1
    
for item in lst:
    print(item, end= ' ')
```

```python
f = open('data.txt', 'r')
f.read() # read all the content of the file at once
f.seek(0)
f.readline(), f.readline(), f.readline() # read content one line at a time
f.seek(0)
# read one line at a time using iteration protocol's next() method
f.__next__(), f.__next__(), f.__next__()

# let python do all the index management of a file using file object
# open built in function generates a file object(iterator)

for line in open('data.txt', 'r'):
    print(line, end='')
```

```python
f = open('data.txt', 'r')
def file_line_generator(file): 
    line = file.readline()
    while line != '':
        yield line
        line = file.readline()

for line in file_line_generator(f):
    print(line, end='')
    

    
```

```python
# automated iteration protocol in action 
for x in [1,2,3]:
    print(x * 10, end=' ')
```

```python
# manual iteration protocol in action
l = [1, 2, 3]
I = iter(l)
while True:
    try:
        item = next(I)
    except StopIteration:
        break
    print(item, end=' ')
```

```python
# iterators are not containers like list, tuples, dict etc
# iterators are on demand data production one data at a time
# iterators can be designed to form chains becuase they are not 
# containers
from itertools import count

numbers = count(100)

f = filter(lambda x: x % 2 == 0, numbers)

m = map(lambda x: x + 2, f)

for n in m:
    print(n)
    if n > 150:
        break
```

```python
class CustomEnum:
    def __init__(self, data):
        self.data = data 
        self.start = 0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.start < len(self.data):
            e = (self.start, self.data[self.start])
            self.start += 1
            return e
        raise StopIteration
        
x = CustomEnum(['apple', 'orange', 'banana'])
x.data
```

```python
for index, item in x:
    print(index, item)
```

```python
# design a custom iterable
class CustomIterable:
    def __init__(self, obj):
        self.obj = obj
        
    def __iter__(self):
        return iter(self.obj)
    
    def __len__(self):
        return len(self.obj)
    
    def __getitem__(self, index):
        return self.obj[index]
    
    def __repr__(self):
        return f'I am {self.obj}'
    
    
x = CustomIterable('hello') # create the instance of the iterable
I = iter(x) # create iterator out of the iterable
x, I, x[:2]

# # manually loop through the iterable object
# counter = 0
# while counter < len(x):
#     print(x[counter])
#     counter += 1

# # loop through the iterator
# for i in I:
#     print(i)
```
