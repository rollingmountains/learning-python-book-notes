```python
# function is a free floating callable code block
def shout(text):
    return text.upper()
shout('this is a beautiful world!')
```

```python
# a method is a function that has a subject and knows how it's working with
class greetings:
    def shout(self):
        return self.text.upper()
    
    def repeat(self, times):
        return self.text * times
    
x = greetings()
x.text = "it's raining outside "
x.shout(), greetings.shout(x)
x.repeat(5), greetings.repeat(x, 5)
```

```python
# def isn't a statemtn but executable code
x = 2
if x == 1:
    def zes():
        print("True")
    x = zes()
    x
else:
    def jes():
        print("False")
    x = jes
    print(x)


```

```python
# use time module to test out function's default synchronous behaviour
import time

def outer():
    print('outer start')
    def slowdown():
        print('inner start')
        time.sleep(5)
        print('inner end')
    print('outer end')
slowdown()
```
