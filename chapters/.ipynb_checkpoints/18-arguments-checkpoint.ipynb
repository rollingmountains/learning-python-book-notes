```python
# positional arguments passing by assignment
def func(c, b, a):
    a += (b + c)
    return a + b + c

x, y, z = 10,20,30

# number of args should match number of params in func header
func(1,2,3) 
```

```python
# keyword arguments passing by assignment
func(c=10, a=30, b=50)
```

```python
# default passing by assignment 
# this is in the definition scope
def func(c, a=10, b=30):
    return a + b + c

# default assignment in header gives the caller an option to skip
# passing the matching arg and it would still work with defaul fallback
func(10)
func(30, 40, 50) # assigns caller's object when passed
```

```python
# starred positional collectors at the function header
def func(x, y, *rest): # header expects two required params 
    # and 1 or more extra optional params
    m = x + y
    j = rest
    return m, j
func(100, 20,400,'test')
```

```python
i, j, *rest = 10, 20
i, j, rest
```

```python
# starred keyword collecter at function header
def func(x,y,**items):
    return x, items
func(x=20, y=30, k='john', l='doe')
```

```python
# pass positional arguments as unpacked collection
l = [1, 'test', 100, 'best']
def func(w,x,*rest):
    return w,x,rest
func(*l)

```

```python
# pass positional arguments by unpacking nested collection
l = [{'name':'john'}, [1,[2,3]], 'test']
func(*l)
```

```python
# pass keyword arguments by unpacking a collection
d = {'name': 'john', 'last':'doe', 'age':40}
def func(age, name, last):
    message = f'Hello {name} {last}. You are {age} years old'
    return message
func(**d)
```

```python
# positional in order from left to right
def func(a, b=99, c=88):
    return a, b, c

func(1,2,3)
```

```python
# can mix positional and keyword but all positional must come before
# keyword
func(10, c=30, b=40)
```

```python
# defaults are for optional arguments
# defaults must follow non defaults because keepin default before
# introduces ambiguity
func(10) # 10, 99, 88 default for b and c
func(10, 20) # 10, 20, 88 defaut for c alone
func(10, 20, 30) # 10, 20, 30 no defaults shared reference with caller
```

```python
# wrapper function that takes in a function, args and returns
# function ready to be called 

def wrapper(func, *args, **kargs):
    print(f'{func.__name__} is running!')
    return func(*args, **kargs)

def add(*args, **kargs): 
    if args:
        total = sum(args)
    elif kargs:
        total = sum([value for value in kargs.values()])
    return total


wrapper(add, **{'a':1, 'b':2, 'c':3})
```

```python
# function as first class citizen object
def add(*add_iterable):
    return sum(add_iterable)

def multiply(*mul_iterable):
    total = 1
    for item in mul_iterable:
        total = total * item
    return total

def exit_instruction():
    print('Thank you. Come again!')

args = [1,9,3]
dispatch = {
    '+': add,
    '*': multiply
}

commands = {
    'exit': exit_instruction
}


while True:
    choice = input('Enter the operator or enter "exit" to end program: ')
    
    cmd = commands.get(choice)
    
    if cmd:
        cmd()
        break
   
    
    func = dispatch.get(choice)
    
    if func is None:
        print('Please enter a valid operator')
        continue  

    print(func(*args))
    print('Enter exit to end the program')
    



    
```

```python
# domain behaviour function
def do_len(text):
    return len(text)

def do_lower(text):
    return text.lower()

def do_upper(text):
    return text.upper()

def do_title(text):
    return text.title()

# meta command behaviours
def exit_instruction():
    print('Thank you. Come again!')

# domain behaviour or action lookup table
text_ops = {
    'len': do_len,
    'upper': do_upper,
    'lower': do_lower,
    'title': do_title
}

# meta behaviour or command lookup table
commands = {
    'exit': exit_instruction }

while True:
    # user input
    user_text = input('Please enter the string: ')
    command = commands.get(user_text)

    if command:
        command()
        break
        
    user_operation = input('Please enter the operation: ')
    
    operation = text_ops.get(user_operation)

    if operation is None:
        print('Please enter a valid operation')
        continue

    print(operation(user_text))





    
```

```python
# find minimum in a list of objects
comparison_ops = {
    '<': lambda x, y:  x < y,
    '>': lambda x, y:  x > y
}

def min1(test, *args):
    res = args[0]
    for arg in args:
        if comparison_ops[test](arg, res):
            res = arg
    return res

def min2(first, *args):
    for arg in args:
        if arg < first:
            first = arg
    return first

def min3(*args):
    temp = list(args)
    temp.sort()
    return temp[0]

def min4(*args):
    return sorted(args)[0]
    
```

```python
def intersection(*iterables):
    result = []
    for item in iterables[0]:
        if item not in result:
            for other in iterables[1:]:
                if item in other:
                    result.append(item)
    return result

intersection([1,2,3,3], [3,4,5])
```
