```python
# nested loops
items = [1, 2, 3, (4,5), 3]
keys = [(4,5), 3.14, 3]

for key in keys:
    for item in items:
        if item == key:
            print(f'{key} found')
            break
    else:
        print(f'{key} not found')

```

```python
# use the membership in to remove the need for inner loop
for key in keys:
    if key in items:
        print(key, 'found')
    else:
        print(key, 'not found')
```

```python
# membership in is still O(n) operation
# to truly move to O(1) use sets which under the hood uses hash table which is O(1)

import timeit



for key in keys:
    if key in set(items):
        print(key, 'found')
    else:
        print(key, 'not found')

        
for key in keys:
    for item in items:
        if item == key:
            print(f'{key} found')
            break
    else:
        print(f'{key} not found')
```

```python
# for loops internal construct
iterable = [1,2,3]
for target in iterable:
    print(target)
    
_iterable = iter(iterable)

while True:
    try:
        target = next(_iterable)
        print(target)
    except StopIteration:
        print('end of loop')
        break
    
```

```python
# using length of sequence items as the list of offsets to work with the item
s = 'hello world'
l = len(s)

for ch in range(l):
    print(ch, end=' ') # prints the offset
print('\n')
for ch in range(l):
    print(s[ch], end='') # prints the character with help of offset
print('\n')
for ch in s:
    print(ch, end='') # works with items, python manages index counter
```

```python
# specific use case for using range and len function together range(len(sequence))
# 1. reordering 2. rotation 3. direct in place changes

# without using range/len, implicit and only can be destructive
s = 'hack'
for _ in s:
    s = s[1:] + s[:1]
    print(s, end=' ')

# using range/len, explicit and give option for both destructive and non destructive changes
print('\n ')
for i in range(len(s) - 1):
    s = s[1:] + s[:1]
    print(s, end=' ')
    
print('\n')
s = 'hack'
for i in range(len(s)):
    s = s[1:i+1] + s[0] + s[i+1:] 
    print(s, end=' ')
```

```python
# traverse a list sequence and change the list as traversing happens
l = [1,2,3,4,5]

for item in l:
    item += 1 # updates only the last of the item and that's not reflected in l
              # this is simply because item is reassigned each time and is actually not inside a container to hold all the values

# this is a good candidate for utilizing access to index using range and len of sequence
for i in range(len(l)):
    l[i] += 1 # incremements all the items by 1

# while loop can also achieve the same result but needs extra counter to keep track of the index
l = [10,20,30,40]
i = 0 # external counter

while i < len(l):
    l[i] += 1
    i += 1 # manual increment of the counter
```

```python
# use list comprehension to achieve traverse and change
x = [x + 1 for x in [10,20,30,40]]
x
```

```python
# prallel traversal using zip() function
l1 = [1,2,3]
l2 = [4,5,6]

p = zip(l1, l2)

# access value pair using iterator protocols next() function
# next(p), next(p), next(p)

# access value using for loop
for (x, y) in p:
    print(f'l1, l2 -> {x}, {y}')
    
# use while loop and index to access the pairs without zip
i = -1
while (i := i + 1) < len(l1):
    print(l1[i], l2[i]) 
    
p = zip(l1, l2)   
# using list comprehension 
x = [(x, y) for x, y in p] # most simplest way to pair
x


```

```python
# zip and dictionaries
# zip is effective when the system awaits the sequence from user or external system, and can only be known during the runtime
keys = ['name', 'age', 'hobby']
values = ['john', 'doe', ['dance', 'reading', 'music']]

# using for loop to construct a dictionary
d = {}
for k, v in zip(keys, values):
    d[k] = v

# using the dict construct directly
d1 = dict(zip(keys, values))

# using dict comprehension
d2 = {k: v for k, v in zip(keys, values)}

d, d1, d2


```

```python
# zip truncate 
l1 = [1,2,3,4,5,6]
l2 = 'hack'
for x, y in zip(l1, l2):
    print(x, y) # the last char in l2 will end up with no pair
    
```

```python
# to work with uneven length of the sequences use itertools
import itertools

l1 = [1,2,3,4,5]
l2 = ['one', 'two', 'three']

for x, y in itertools.zip_longest(l2, l1):
    print(x, y)
```

```python
# enumerate function gives access to both the index and items as pairs, as usual enumerate is a lazy iterator that returns or yields a pair of index, item in tuple form
x = enumerate('hack')
next(x), next(x), next(x), next(x) # -> ((0, 'h'), (1, 'a'), (2, 'c'), (3, 'k'))

for index, item in enumerate('hello'):
    print(item, 'has offset at', index)
```

```python
# make lazy iterator into eager iterator by capturing it in a container
x = [(index, item) for index, item in enumerate('world', start=100)]
x
```
