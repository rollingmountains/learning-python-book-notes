```python
# function returning a function
def power(exponent, test, message ):
    def raise_to(x):
        return f'{x ** exponent + test} + {message}'
    return raise_to

square = power(2, 100, 'i am here')
print(square.__closure__)
print(square.__closure__[0].cell_contents)
print(square.__closure__[1].cell_contents)
print(square.__closure__[2].cell_contents)
result = square(10)
result
```

```python
# pure basic function 
def expo(x, exponent):
    return x ** exponent

x, y = expo(10, 2), expo(10, 3)
x,y
```

```python
# use closure to acheive the same above 
def power(exponent):
    def raise_to(x):
        return x ** exponent
    return raise_to

square, cube = power(2), power(3)
square(10), cube(10)
```

```python
# use class object model to achieve the same
class Exponent:
    def __init__(self, expo):
        self.expo = expo
        
    def raise_to(self, x):
        return x ** self.expo
    
square, cube = Exponent(2), Exponent(3)
square.raise_to(10), cube.raise_to(10)
    
```

```python
class CreditCard:
    def __init__(self, amount):
        self.name = 'Credit Card'
        self.amount = amount
        
    def pay(self):
        return f'Payment {self.amount} done using {self.name}'

class Stripe:
    def __init__(self, amount):
        self.name = 'Stripe'
        self.amount = amount
        
    def pay(self):
        return f'Payment {self.amount} done using {self.name}'
    
class Paypal:
    def __init__(self, amount):
        self.name = 'Paypal'
        self.amount = amount
        
    def pay(self):
        return f'Payment {self.amount} done using {self.name}'
    
def payment(gateway):
    return gateway.pay()
    
payment(CreditCard(200)), payment(Stripe(300)), payment(Paypal(400))
```

```python
s1 = [1,2,3]
s2 = [2,1,3]
def intersect(seq1, seq2):
    res = []
    for x in seq1:
        for y in seq2:
            if x == y:
                res.append(x)
    return res
x = intersect(s1, s2)
x
```

```python
def intersect(seq1, seq2):
    res = [x for x in seq1 for y in seq2 if x == y]
    return res
intersect(s1, s2)
```

```python
def f1(x):
    return x * 2


def f2(x):
    return str(x) + "!"


def chain(funcs, val):
    for f in funcs:
        val = f(val)
    return val

result = chain([f1, f2, f1], 3)
chaining = f1(f2(f1(3)))
result, chaining

```
