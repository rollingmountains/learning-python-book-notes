```python
# find max value 
def max_value(nums):
  if not nums:
    return None
  # container 
  largest_number = nums[0]

  # filter/comparison
  for item in nums[1:]:
    if item > largest_number:
      largest_number = item

  return largest_number

max_value([]), max_value([-1,0,2,99])
```

```python
# find the longest word
def longest_word(sentence):
  if not sentence:
    return 'sentence is empty'

  split_sentence = sentence.split(' ')

  longest = split_sentence[0]

  for word in split_sentence[1:]:
    if len(word) >= len(longest):
      longest = word

  return longest 

longest_word(''), longest_word('potato')
```
