```python
# in memory data
# a list of dictionary objects that hold a record
people = [
    {"name": "Amit", "age": 25, "city": "Kathmandu"},
    {"name": "Sita", "age": 30, "city": "Pokhara"},
    {"name": "Bishal", "age": 22, "city": "Lalitpur"},
    {"name": "Rita", "age": 28, "city": "Kathmandu"},
]



# a function that selects the record form that list when the city = kathmandu
def filter_by_city(**kargs):
    if 'city' not in kargs:
        raise KeyError('Please provide a valid parameter.')
        
    search_city = kargs.get('city').lower()
        
    if not search_city:
        raise ValueError('Parameter city cannot be empty')
    
    
    result = [record for record in people if record.get('city', '').lower() == search_city]
    return result

# lookup for comparison operator
import operator

ops_dict = {
        '>': operator.gt,
        '<': operator.lt,
        '>=': operator.ge,
        '<=': operator.le,
        '==': operator.eq,
        '!=': operator.ne
    }

# fitler by age
def filter_by_age(age, op):
    result = [record for record in people if ops_dict[op](record.get('age', ''), age)]
    return result


# function lookup
make_query = {
    'city': filter_by_city,
    'age': filter_by_age
}

# function call
def run_query(name, *args, **kargs):
    print(f'Dispatcher received args: {args} and {kargs} for {name}')
    return make_query[name](*args, **kargs)


run_query('age', 25, op=">")

# run_query('city', **{'city': 'PoKHAra'})

# filter_by_city(**{'city': 'PoKHAra'})
```
