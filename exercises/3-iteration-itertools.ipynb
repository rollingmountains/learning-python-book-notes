```python
class CustomRange:
    def __init__(self, limit):
        self.limit = limit
        self.index = 0
        
    def __iter__(self):
        return self
        
    def __next__(self):
        if self.index < self.limit:
            value = self.index
            self.index += 1
            return value
        else:
            raise StopIteration 
    
r = CustomRange(3)
next(r), next(r), next(r)
```

```python
class CountDown:
    def __init__(self, count):
        self.count = count
        
    def __iter__(self):
        return self
        
    def __next__(self):
        value = self.count 
        
        if value < 0:
            raise StopIteration
         
        self.count -= 1
        return value

x = CountDown(3)
next(x), next(x), next(x), next(x)
for x in CountDown(5):
    print(x)
```

```python
# multiple context manager
with open('sample.txt') as f1:
    with open('new-sample.txt', 'w') as f2:
        for line in (line.rstrip() for line in f1):
            f2.write(line + '\n')
```
